```python
#import random
#from Character import Character
#from Novice import Novice
#from Archer import Archer
#from Swordsman import Swordsman
#from Magician import Magician
#from Boss import Boss
import random
class Character():
    def __init__(self, username):
        self.__username = username
        self.__hp = 100
        self.__mana = 100
        self.__damage = 5
        self.__str = 0
        self.__vit = 0
        self.__int = 0
        self.__agi = 0
    def getUsername(self):
        return self.__username
    def setUsername(self, new_username):
        self.__username = new_username
    def getHp(self):
        return self.__hp
    def setHp(self, new_hp):
        self.__hp = new_hp
    def getDamage(self):
        return self.__damage
    def setDamage(self, new_damage):
        self.__damage = new_damage
    def getStr(self):
        return self.__str
    def setStr(self, new_str):
        self.__str = new_str
    def getVit(self):
        return self.__vit
    def setVit(self, new_vit):
        self.__vit = new_vit
    def getInt(self):
        return self.__int
    def setInt(self, new_int):
        self.__int = new_int
    def getAgi(self):
        return self.__agi
    def setAgi(self, new_agi):
        self.__agi = new_agi
    def reduceHp(self, damage_amount):
        self.__hp = self.__hp - damage_amount
    def addHp(self, heal_amount):
        self.__hp = self.__hp + heal_amount

class Novice(Character):
    def basicAttack(self, character):
        character.reduceHp(self.getDamage())
        print(f"{self.getUsername()} performed Basic Attack! -{self.getDamage()}")

class Swordsman(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(5)
        self.setVit(10)
        self.setHp(self.getHp()+ self.getVit())

    def slashAttack(self, character):
        self.new_damage = self.getDamage()+self.getStr()
        character.reduceHp(self.new_damage)
        print(f"{self.getUsername()} performed Slash Attack! -{self.new_damage}")

class Archer(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setAgi(5)
        self.setInt(5)
        self.setVit(5)
        self.setHp(self.getHp()+self.getVit())

    def rangedAttack(self, character):
        self.new_damage = self.getDamage()+random.randint(0,self.getInt())
        character.reduceHp(self.new_damage)
        print(f"{self.getUsername()} performed Slash Attack! -{self.new_damage}")

class Magician(Novice):
    def __init__(self, username):
        super().__init__(username)
        self.setInt(10)
        self.setVit(5)
        self.setHp(self.getHp()+self.getVit())

    def heal(self):
        self.addHp(self.getInt())
        print(f"{self.getUsername()} performed Heal! +{self.getInt()}")

    def magicAttack(self, character):
        self.new_damage = self.getDamage()+self.getInt()
        character.reduceHp(self.new_damage)
        print(f"{self.getUsername()} performed Magic Attack! -{self.new_damage}")

class Boss(Swordsman, Archer, Magician):
    def __init__(self, username):
        super().__init__(username)
        self.setStr(10)
        self.setVit(25)
        self.setInt(5)
        self.setHp(self.getHp()+self.getVit())

def select_role(player_name, allow_all_roles=True):
    roles = {"1":Novice, "2":Swordsman, "3":Archer, "4":Magician}
    if not allow_all_roles:
        roles = {"1":Novice}

    print("Select your role: ")
    for key, role in roles.items():
        print(f"{key}.{role.__name__}")

    choice = input(f"Enter your choice: ")
    return roles.get(choice, Novice)(player_name)

def battle(player1,player2):
    players = [player1,player2]
    random.shuffle(players)

    print(f"{players[0].getUsername()} vs. {players[1].getUsername()} - Battle Start!")
    while players[0].getHp()>0 and players[1].getHp()>0:
        attacker, defender = players
        print(f"{attacker.getUsername()} attacks!")
        attacker.basicAttack(defender)
        print(f"{defender.getUsername()} HP: {defender.getHp()}")

        if defender.getHp() <=0:
            print(f"{attacker.getUsername()} wins!")
            return attacker

        players.reverse()

def single_player_mode():
    wins = 0
    player = select_role("Player", allow_all_roles=False)
    while True:
        boss = Boss("Monster")
        winner = battle(player, boss)

        if winner == player:
            wins +=1
            print(f"You have {wins} wins!")
            if wins ==2:
                print("You've won 2 matcher! You can now select a new roles. ")
                player = select_role(player.getUsername())
        else:
            print("Game Over!")
            break

        if input("Play again? (y/n): ").lower() !="y":
            break

def player_vs_player_mode():
    player1 = select_role("Player 1")
    player2 = select_role("Player 2")
    battle(player1, player2)

def main():
    while True:
        print(f"\nSelect Game Mode: ")
        print("1. Single Player")
        print("2. Player vs. Player")
        print("3. Exit")
        choice = input("Enter choice: ")

        if choice == "1":
            single_player_mode()
        elif choice == "2":
            player_vs_player_mode()
        elif choice == "3":
            print("Exiting game.")
            break
        else:
            print("Invalid choice, Try again")

if __name__ == "__main__":
    main()
```
