```python
class Student:
  def __init__ (self, id, name, program, sex, weight, height, course, midterm_cs, midterm_grade, final_cs, final_grade, semestral_grade):
    self.id = id
    self.name = name
    self.program = program
    self.sex = sex
    self.weight = weight
    self.height = height
    self.course = course

    self.midterm_cs = ((midseatwork*0.3) + (midquiz*0.3) + (Mid_ass*0.2) + (midrecitation*0.2))
    self.midterm_grade = ((self.midterm_cs*0.6) + (midterm_exam*0.4))

    self.final_cs = ((finalseatwork*0.3) + (finalquiz*0.3) + (final_ass*0.2) + (finalrecitation*0.2))
    self.final_grade = ((self.final_cs*0.6) + (final_exam*0.4))

    self.semestral_grade = (((0.3)*self.midterm_grade) + ((0.7)*self.final_grade))

  def convert(self):
      if self.semestral_grade >= 97:
          return 1.00
      elif self.semestral_grade >= 94:
          return 1.25
      elif self.semestral_grade >= 91:
          return 1.50
      elif self.semestral_grade >= 88:
          return 1.75
      elif self.semestral_grade >= 85:
          return 2.00
      elif self.semestral_grade >= 82:
          return 2.25
      elif self.semestral_grade >= 79:
          return 2.50
      elif self.semestral_grade >= 76:
          return 2.75
      elif self.semestral_grade >= 60:
          return 3.00
      elif self.semestral_grade >= 50:
          return 4.00
      elif self.semestral_grade >= 40:
          return 5.00
      else:
          return "Failed"

  def description(self):
      print(f"Student id:       {self.id}")
      print(f"Sutdent name:     {self.name}")
      print(f"Student program:  {self.program}")
      print(f"Student sex:      {self.sex}")
      print(f"Student weight:   {self.weight}")
      print(f"Student height:   {self.height}")
      print(" ")
      print(f"Course:           {self.course}")
      print(f"Midterm Grade:    {self.midterm_grade}")
      print(f"Final Grade:      {self.final_grade}")
      print(f"Semestral Grade:  {self.semestral_grade}")
      print(f"Equivalent Grade: {self.convert()}")
      print("-" * 30)

students = []
n = int(input("Enter Number of Students: "))
for x in range(n):
  id = int(input(f"\nEnter Student {X+1} Id: "))
  name = input(f"Enter Student {x+1} Name: ")
  program = str(input("Enter Student Program: "))
  sex = str(input("Enter Student Sex: "))
  weight = float(input("Enter Student Weight(kg): "))
  height = float(input("Enter Student Height(cm): "))
  course = input(f"\nEnter Course: ")

  midseatwork = float(input(f"\nEnter Student {x+1} Midterm Seatwork Grade: "))
  midquiz = float(input(f"Enter Student {x+1} Midterm quiz Grade: "))
  Mid_ass = float(input(f"Enter Student {x+1} Midterm assignment Grade: "))
  midrecitation = float(input(f"Enter Student {x+1} midterm recitation Grade: "))
  midterm_exam = float(input("Enter Student Midterm Exam Score: "))

  finalseatwork = float(input(f"\nEnter Student {x+1} final seatwork: "))
  finalquiz = float(input(f"Enter Student {x+1} final quiz: "))
  final_ass = float(input(f"Enter Student {x+1} final assignment: "))
  finalrecitation = float(input(f"Enter Student {x+1} final recitation: "))
  final_exam = float(input("Enter Student Final Exam Score:  "))

  students.append(Student(id, name, program, sex, weight, height, course, 0, 0, 0, 0,0 ))

print("\nStudent List: ")
for b in students:
    b.description()
```
